```python
class ImplantesMedicos: #CLASE PADRE
    def __init__(self, fechaFabricacion, material, nombre, tamaño):
        self._fechaFabricacion = fechaFabricacion
        self._material = material
        self._nombre = nombre
        self._tamaño = tamaño

    @property
    def fechaFabricacion(self):
        return self._fechaFabricacion
    @fechaFabricacion.setter
    def fechaFabricacion(self, fechaFabricacion):
        self._fechaFabricacion = fechaFabricacion
    @property
    def material(self):
        return self._material
    @material.setter
    def material(self, material):
        self._material = material
    @property
    def nombre(self):
        return self._nombre
    @nombre.setter
    def nombre(self, nombre):
        self.nombre= nombre
    @property
    def tamaño(self):
        return self._tamaño
    @tamaño.setter
    def tamaño(self, tamaño):
        self._tamaño = tamaño

    def mostrarInformacion(self):
        print(f"El dispositivo médico es: {self._nombre}, está elaborado de: {self._material}\n y tiene Fecha de fabricación: {self._fechaFabricacion}")

class ImplanteCadera(ImplantesMedicos):
    def __init__(self,material,  tamaño, tipoFijacion):
       super().__init__(material, tamaño, tipoFijacion)
       self.__tipoFijacion= tipoFijacion

    @property
    def tipoFijacion(self):
        return self.__tipoFijacion
    @tipoFijacion.setter
    def tipoFijacion(self, tipoFijacion):
        self.__tipoFijacion = tipoFijacion

    def __str__(self):
        return f'{super().__str__()}, {self.__tipoFijacion}'

class ImplanteRodilla(ImplantesMedicos):
    def __init__(self,material,  tamaño, tipoFijacion):
       super().__init__(material, tamaño)
       self.__tipoFijacion= tipoFijacion

    @property
    def tipoFijacion(self):
        return self.__tipoFijacion
    @tipoFijacion.setter
    def tipoFijacion(self, tipoFijacion):
        self.__tipoFijacion = tipoFijacion

    def __str__(self):
        return f'{super().__str__()}, {self.__tipoFijacion}'


class StentCoronario(ImplantesMedicos):
    def __init__(self, fechaFabricacion, material, nombre, tamaño, longitud, diametro):
        super().__init__( fechaFabricacion, material, nombre, tamaño)
        self.longitud=longitud
        self.diámetro=diametro
    @property
    def longitud(self):
        return self.longitud
    @longitud.setter
    def longitud(self,longitud):
        self.longitud=longitud

    @property
    def diametro(self):
        return self.diametro
    @diametro.setter
    def diametro(self,diametro):
        self.diametro=diametro

    def __str__(self):
         return f'{super().__str__()}, {self.longitud}, {self.diametro}'


class ImplanteDental(ImplantesMedicos):
    def __init__(self, fechaFabricacion, material, nombre, tamaño, forma, sistemaFijacion):
        super().__init__(fechaFabricacion, material, nombre, tamaño)
        self.forma=forma
        self.sistemaFijación=sistemaFijacion

    @property
    def forma(self):
        return self.forma
    @forma.setter
    def forma(self,forma):
        self.forma=forma

    @property
    def sistemaFijacion(self):
        return self.sistemaFijación
    @sistemaFijacion.setter
    def sistemaFijacion(self,sistemaFijacion):
        self.sistemaFijación=sistemaFijacion

    def __str__(self):
         return f'{super().__str__()}, {self.forma}, {self.sistemaFijacion}'

class Marcapasos(ImplantesMedicos):
    def _init_(Self, fechaFabricacion, material, nombre, tamaño, cantidadElectrodos, alambricoOinalambrico, frecuenciaEstimulacion):
        super().__init__(fechaFabricacion, material, nombre, tamaño, cantidadElectrodos, alambricoOinalambrico, frecuenciaEstimulacion)
        self.cantidadElectrodos=cantidadElectrodos
        self.alambricoOinalambrico=alambricoOinalambrico
        self.frecuenciaEstimulacion=frecuenciaEstimulacion

    @property
    def cantidadElectrodos(self):
        return self.cantidadElectrodos
    @cantidadElectrodos.setter
    def cantidadElecctrodos(self,cantidadElectrodos):
        self.cantidadElectrodos=cantidadElectrodos

    @property
    def alambricoOinalambrico(self):
        return self.alambricoOinalambrico
    @alambricoOinalambrico.setter
    def alambricoOinalambrico(self,alambricoOinalambrico):
        self.alambricoOinalambrico=alambricoOinalambrico

    @property
    def frecuenciaEstimulacion(self):
        return self.frecuenciaEstimulacion
    @frecuenciaEstimulacion.setter
    def frecuenciaEstimulacion(self,frecuenciaEstimulacion):
        self.frecuenciaEstimulacion=frecuenciaEstimulacion

    def __str__(self):
         return f'{super().__str__()}, {self.cantidadElectrodos}, {self.alambricoOinalambrico}, {self.frecuenciaEstimulacion}'

class Paciente:
    def __init__(self, nombre, edad, documento):
        self.nombre = nombre
        self.edad = edad
        self.id_paciente =documento
        self.asignarImplantes = [] # lista en donde se me van a almacenar los implantes asignados al respectivo paciente

    def asignar_implante(self, implante):
        self.asignarImplantes.append(implante) # en mi lista se va a guardar el implante asignado

    def __str__(self):
        return f"Nombre: {self.nombre}, Edad: {self.edad}, ID: {self.documento}"

class AsignacionImplante:
    def __init__(self, paciente, implante, fecha_implantacion, medico_responsable):
        self.paciente = paciente
        self.implante = implante
        self.fecha_implantacion = fecha_implantacion
        self.medico_responsable = medico_responsable
    def __str__(self):
        return f"Paciente: {self.paciente.nombre}, Implante: {self.implante.nombre}, Fecha: {self.fecha_implantacion}, Médico: {self.medico_responsable}"

class InventarioImplantes:
    def __init__(self):
        self.implantes = [] # nuestra lista se llama implantes
    def agregar_implante(self, implante):
        self.implantes.append(implante) # se va a agregar el implante para el inventario
    def eliminar_implante(self, nombre_implante): #cambiamos el metodo de elimianr implante porque nos dimos cuenta que podia ocurrir un erros, ya que si hay más de un implante con el mismo nombre, se eliminarán todos
      for implante in self.implantes:
        if implante.nombre==nombre_implante:
          self.implantes.remove(implante)
          break

    def visualizar_inventario(self):
        for implante in self.implantes: #ciclo for que permite iterar sobre cada elemento de implante en la lista self.implantes
            print(implante.mostrarInformacion())

class RegistroPacientes:
    def __init__ (self):
      self.pacientes={} #diccionario vacio
    def rergistrarPaciente(self,paciente, implante):
        if paciente not in self.pacientes:
          self.pacientes[paciente]=[implante]
        else:
          self.pacientes[paciente].append(implante)
    def asignarImplante (self,paciente,implante):
        self.pacientes[paciente]=implante
    def VisualizarPacientes(self):
        for paciente, implantes in self.pacientes.items ():
          print(f"paciente: {paciente}")
          for implante in implantes:
            print(f"imlante: {implante.mostrarInformacion()}")

class SeguimientoImplantes:
    def __init__(self):
        self.implantes_seguimiento = {}

    def agregarSeguimiento(self, implante, fechaRevision, estado):
        if implante not in self.implantes_seguimiento:
            self.implantes_seguimiento[implante] = [(fechaRevision, estado)]
        else:
            self.implantes_seguimiento[implante].append((fechaRevision, estado))

    def visualizarSeguimiento(self):
        for implante, seguimiento in self.implantes_seguimiento.items():
            print(f"Implante: {implante.mostrarInformacion()}")
            for fecha, estado in seguimiento:
                print(f"Fecha de revisión: {fecha}, Estado: {estado}")

    def editar_implante(self):
        nombre_implante = input("por favor ingrese el implante que desea modificar: ")
        implantes = [implante for implante in self.inventario.implantes if implante.nombre == nombre_implante]
        if not implantes:
            print("Señor usuario hay un error, no se encuentra el implante a modificar .")
            return

        implante = implantes[0]
        print("Implante encontrado. Ingrese los nuevos datos:")
        fecha_fabricacion = input("Fecha de fabricación: ")
        material = input("Material: ")
        tamaño = input("Tamaño: ")
        implante.fechaFabricacion = fecha_fabricacion
        implante.material = material
        implante.tamaño = tamaño

        print("Información del implante actualizada.")

class InterfazUsuario:
    def __init__(self, inventario):
        self.inventario = inventario

    def mostrar_menu(self):
        while True:
            menu= input(""" Bienvenido al sistema para el manejo de implantes medicos,
                        Elija una de las siguientes opciones:
                         1.Agregar nuevos implantes
                         2.Eliminar implantes
                         3.Editar informacion de implantes
                        4.Visualizar inventario completo
                        5.Salir
                        """ )
            if menu== "1":
                self.agregar_implante()
            elif menu== "2":
                self.eliminar_implante()
            elif menu== "3":
                self.editar_implante()
            elif menu== "4":
                self.visualizar_inventario()
            elif menu== "5":
                break
            else:
                   print("OPCION NO VALIDA. INTENTE NUEVAMENTE")
    def agregar_implante(self):
        nombre = input("Ingrese el nombre del implante: ")
        fecha_fabricacion = input("Ingrese la fecha de fabricación: ")
        material = input("Ingrese el material del implante: ")
        tamaño = input("Ingrese el tamaño del implante: ")

        nuevo_implante = ImplantesMedicos(fecha_fabricacion, material, nombre, tamaño)
        self.inventario.agregar_implante(nuevo_implante)
        print("Implante agregado al inventario.")

    def eliminar_implante(self):
        nombre_implante = input("Ingrese el nombre del implante que desea eliminar: ")
        self.inventario.eliminar_implante(nombre_implante)
        print("Implante eliminado del inventario.")

    def editar_implante(self):
        nombre_implante = input("Ingrese el nombre del implante que desea editar: ")
        implantes = [implante for implante in self.inventario.implantes if implante.nombre == nombre_implante]
        if not implantes:
            print("Implante no encontrado.")
            return

        implante = implantes[0]
        print("Implante encontrado. Ingrese los nuevos datos:")
        fecha_fabricacion = input("Fecha de fabricación: ")
        material = input("Material: ")
        tamaño = input("Tamaño: ")
        self.inventario.eliminar_implante(nombre_implante)
        nuevo_implante=ImplantesMedicos(fecha_fabricacion,material,nombre_implante, tamaño)
        self.inventario.agregar_implante(nuevo_implante)
        print("Información del implante actualizada")

    def visualizar_inventario(self):
        self.inventario.visualizar_inventario()






```
