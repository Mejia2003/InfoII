```python
#INTEGRANTES DEL GRUPO:

#ALMA OLEA BARBOSA 
#NICOLLE MEJIA 

#PUNTO 1


import cv2
import matplotlib.pyplot as plt
import numpy as np

class ProcesadorImagen:

    def _init_(self, ruta_imagen):
        self.imagen_original = self.cargar_imagen(ruta_imagen)
        self.imagen_umbrelizada = None

    
    def cargar_imagen(self, ruta):
        imagen = cv2.imread(ruta)
        if imagen is None:
            raise ValueError(f"No se pudo cargar la imagen en la ruta: {ruta}")
        print(f"Imagen cargada con éxito. Forma: {imagen.shape}")
        return imagen

    def mostrar_imagen(self, imagen, titulo="imagen celulas"):
        if len(imagen.shape) == 2:  # Comprobar si la imagen es en escala de grises
            plt.imshow(imagen, cmap='gray')  # Mostrar la imagen en escala de grises
        else:
            plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))  # Mostrar la imagen en color
        plt.title(titulo)
        plt.axis('off')
        plt.show()

    def umbrellizar(self):
        imagen_gris = cv2.cvtColor(self.imagen_original, cv2.COLOR_BGR2GRAY)
        _, self.imagen_umbrelizada = cv2.threshold(imagen_gris, 120, 255, cv2.THRESH_BINARY)



    def dilatar_imagen(self, iterations=1):
        self.imagen_umbrelizada = cv2.dilate(self.imagen_umbrelizada, np.ones([5, 5]), iterations=iterations)
        self.mostrar_imagen(self.imagen_umbrelizada, "Imagen dilatada")

    def erosionar(self, iterations=1):
        self.imagen_umbrelizada = cv2.erode(self.imagen_umbrelizada, np.ones([5, 5]), iterations=iterations)
        self.mostrar_imagen(self.imagen_umbrelizada, "Imagen erosionada")

    def encontrar_contornos(self):
        _, contornos, _ = cv2.findContours(self.imagen_umbrelizada, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        return len(contornos)

    def inicializar(self):
        self.umbrellizar()
        self.mostrar_imagen(self.imagen_original, "Imagen Original")
        self.mostrar_imagen(self.imagen_umbrelizada, "Imagen umbralizada")

    def dibujar_contornos(self):
        imagen_dibujada = self.imagen_original.copy()
        contornos, _ = cv2.findContours(self.imagen_umbrelizada, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        numero_celulas = 0
        for contorno in contornos:
            area = cv2.contourArea(contorno)
            if area > 100:
                numero_celulas += 1
                cv2.drawContours(imagen_dibujada, [contorno], -1, (0, 255, 0), 2)
        self.mostrar_imagen(imagen_dibujada, "Imagen con contornos")
        return numero_celulas
```

```python
procesador = ProcesadorImagen(r"\Users\nidam\InfoII\taller2_monitor\Imagencelula.jpg")
procesador.mostrar_imagen(procesador.imagen_original, "Imagen Original")
procesador.umbrellizar()
procesador.mostrar_imagen(procesador.imagen_umbrelizada, "Imagen Umbralizada")
procesador.dilatar_imagen()
procesador.erosionar()
numero_celulas = procesador.dibujar_contornos()
print(f"Número de células encontradas: {numero_celulas}")
```

```python
#PUNTO 2

import os
import pydicom
import matplotlib.pyplot as plt

ruta= r"C:\Users\USUARIO\Documents\REPOS\Info-2\archivosDCM"

def cargar_imagen(ruta):
    archivos= [file for file in os.listdir(ruta) if file.endswith('.dcm')]
    archivos.sort()
    imagen= [pydicom.dcmread(os.path.join(ruta,file)) for file in archivos]
    return imagen 
    
def mostrar_imagen(imagen):
    for im in imagen: 
        plt.imshow(im.pixel_array, cmap='Blues')
        plt.title(f'Imagen: {os.path.split(im.filename)[-1]}')  
        plt.axis('off')
        plt.colorbar()
        plt.show()


imagenes = cargar_imagen(ruta)
mostrar_imagen(imagenes)
```
